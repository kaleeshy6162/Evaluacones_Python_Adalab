```python
# Los nombres de las columnas empiezan con mayúsculas, pon todos los nombres de las columnas en minúsculas.
columnas_nuevas = []
for columna in df.columns:
    columnas_nuevas.append(columna.lower())

df.columns = columnas_nuevas
df.columns
```

```python
# Los nombres de las columnas empiezan con mayúsculas, pon todos los nombres de las columnas en minúsculas.
df.columns = [col.lower() for col in df.columns]
df.columns
```

```python
# Crea una función que nos permita cambiar esas comas por puntos para que los datos tengan el tipo correcto. (danceability, energy, key, loudness, speechiness, acousticness, instrumentalness)

def corregir_columnas_comas_a_puntos(columnas):
    """
    Corrige los valores en las columnas especificadas, reemplazando comas por puntos 
    y convirtiendo los valores a float.
    pf_corregido: El DataFrame con las columnas corregidas.
    """
    try:
        if isinstance(columnas, str):  # Esta línea verifica si columnas es una cadena
            columnas = [columnas]   
            
        for columna in columnas:
            if columna in df.columns:
                df[columna] = df[columna].astype(str).str.replace(',', '.').astype(float)
                
            else:
                print(f"Advertencia: La columna '{columna}' no está en el DataFrame.")
        return df
    
    except Exception as e:
        print(f"Ocurrió un error al procesar las columnas: {e}")
        return df
```

```python
# Lista de columnas a corregir
columnas_a_corregir = ['danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness']

# Aplicar la función al DataFrame
df_corregido = corregir_columnas_comas_a_puntos(columnas_a_corregir)
```
